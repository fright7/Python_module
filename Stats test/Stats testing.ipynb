```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
class Stats_Test:
    
    def __init__(self, array_x,alternative, array_y=None, val=None, var=None,alpha=None, normal=False):
    
        
        self.array_x = array_x
        self.array_y = array_y
        self.val = val
        self.var = var
        self.normal = normal
        self.alternative = alternative
        self.alpha = alpha
        
    
    
    def test(self):
        
        
        if (self.array_y == None) & (self.val== None):
            
            print("The test cannot be conducted due to empty value")
        
        
        elif (self.array_y == None) & (self.val!=None):
            
            if self.normal == False:
                
                if len(self.array_x) >= 30:
                    
                    if self.var == None:
                        
                        return self.t_test2()
                    
                    else:
                        
                        return self.z_test1()
                    
                elif len(self.array_x) < 30:
                    
                    print("Warning : The sample size is too small")
                    
                    if self.var == None:
                        
                        return self.t_test1()
                    
                    else:
                        
                        return self.z_test1()
                    
            elif self.normal == True:
                
                if len(self.array_x) >= 30:
                    
                    if self.var == None:
                        
                        return self.t_test2()
                    
                    else:
                        
                        return self.z_test1()
                    
                elif len(self.array_x) < 30:
                    
                    if self.var == None:
                        
                        return self.t_test1()
                    
                    else:
                        
                        return self.z_test1()
        
        else:
            
            if self.val == None:
                
                self.val = 0
            
            
            if self.normal == False:
                
                if (len(self.array_x) >= 30) & (len(self.array_y) >=30):
                    
                    if self.var == None:
                        
                        return self.t_test5()
                    
                    else:
                        
                        return self.z_test2()
                    
                else:
                    
                    print("Warning : The sample sizes are too small")
                        
                    if self.var == None:
                            
                        if self.alpha == None:
                            
                            print("Warning : Input Alpha")
                        
                        
                        else:
                            
                            if self.f_test() == True:
                                
                                return self.t_test4()
                            
                            else:
                                
                                return self.t_test3()
                            
                    else:
                            
                        return self.z_test2()
            else:
                    
                    if (len(self.array_x) >= 30) & (len(self.array_y) >=30):
                    
                        if self.var == None:
                        
                            return self.t_test5()
                    
                        else:
                        
                            return self.z_test2()
                        
                    else:
                        
                        if self.var == None:
                            
                            if self.alpha == None:
                                
                                print("Warning : Input Alpha")
                            
                            else:
                                
                                if self.f_test() == True:
                                
                                    return self.t_test4()
                            
                                else:
                                
                                    return self.t_test3()
                            
                        else:
                            
                            return self.z_test2()
                
    
    def z_test1(self): # x z test
        
        print("z1")
        z = ((np.mean(self.array_x) - self.val)) / (self.var/np.sqrt(len(self.array_x)))
        
        if self.alternative == "neq":
             
            p_val = 2*(1-stats.norm.cdf(abs(z)))
            return p_val
        
        elif (self.alternative == ">"): 
            
            p_val = (1-stats.norm.cdf(z))
            return p_val
        
        else:
            
            p_val = stats.norm.cdf(z)
            return p_val
        
    def t_test1(self): # x t test
        
        print("t1")
        
        samp_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        
        t = (np.mean(self.array_x) - self.val) / (np.sqrt(samp_var/len(self.array_x)))
        
        if self.alternative == "neq":
            
            p_val = 2*(1-stats.t.cdf(abs(t),df=len(self.array_x)-1))
            return p_val
        
        elif (self.alternative == ">"): 
            
            p_val = 1-stats.t.cdf(t,df=len(self.array_x)-1)
            return p_val
        
        else:
            
            p_val = stats.t.cdf(t,df=len(self.array_x)-1)
            return p_val
        
    def t_test2(self): # x normal false, central
        
        print("t2")
        
        samp_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        
        t = (np.mean(self.array_x) - self.val) / (np.sqrt(samp_var/len(self.array_x)))
        
        if self.alternative == "neq":
            
            p_val = 2*(1-stats.norm.cdf(abs(t)))
            return p_val
            
        
        elif (self.alternative == ">"): 
            
            p_val = 1-stats.norm.cdf(t)
            return p_val
        
        else:
            
            p_val = stats.norm.cdf(t)
            return p_val
            
        
            
    def z_test2(self): # x & y z test
        
        print("z2")
        
        z = (np.mean(self.array_x) - np.mean(self.array_y) -self.val)/(np.sqrt(self.var[0]/len(self.array_x) + self.var[1]/len(self.array_y)))
        
        if self.alternative == "neq":
             
            p_val = 2*(1-stats.norm.cdf(abs(z)))
            return p_val
        
        elif (self.alternative == ">"):
            
            p_val = (1-stats.norm.cdf(z))
            return p_val
        
        else:
            
            p_val = stats.norm.cdf(z)
            return p_val
        
    
    def t_test3(self):
        
        # x std = y std t test
        
        print("t3")
        
        samp1_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        samp2_var = np.var(self.array_y)*len(self.array_y)/(len(self.array_y)-1)
        
        s_var = ((len(self.array_x)-1)*samp1_var + (len(self.array_y)-1)*samp2_var) / (len(self.array_x) + len(self.array_y)-2)
        
        t = (np.mean(self.array_x)-np.mean(self.array_y) - self.val) / (np.sqrt(s_var)*np.sqrt(1/len(self.array_x)+1/len(self.array_y)))
        
        
        if self.alternative == "neq":
            
            p_val = 2*(1-stats.t.cdf(abs(t),df=len(self.array_x)+len(self.array_y)-2))
            return p_val
        
        elif (self.alternative == ">"):
            
            p_val = 1-stats.t.cdf(t,df=len(self.array_x)+len(self.array_y)-2)
            return p_val
        
        else:
            
            p_val = stats.t.cdf(t,df=len(self.array_x)+len(self.array_y)-2)
            return p_val
        
    def t_test4(self):
        
        # x std != y std t test
        
        print("t4")
        
        samp1_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        samp2_var = np.var(self.array_y)*len(self.array_y)/(len(self.array_y)-1)
        
        nom = (samp1_var/len(self.array_x)+samp2_var/len(self.array_y))**2
        denom = (samp1_var/len(self.array_x))**2 / (len(self.array_x)-1) + (samp2_var/len(self.array_y))**2 / (len(self.array_y)-1)
        
        df = nom / denom
        
        t = (np.mean(self.array_x)-np.mean(self.array_y)-self.val) / (np.sqrt(samp1_var / len(self.array_x) + samp2_var / len(self.array_y)))
        
        if self.alternative == "neq":
            
            p_val = 2*(1-stats.t.cdf(abs(t),df=df))
            return p_val
        
        elif (self.alternative == ">"): 
            
            p_val = 1-stats.t.cdf(t,df=df)
            
            return p_val
        
        else:
            
            p_val = stats.t.cdf(t,df=df)
            
            return p_val
    
    def t_test5(self):
        
        # normal=False, central limit
        
        print("t5")
        samp1_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        samp2_var = np.var(self.array_y)*len(self.array_y)/(len(self.array_y)-1)
        
        t = (np.mean(self.array_x) - np.mean(self.array_y) - self.val) / np.sqrt(samp1_var/len(self.array_x) + samp2_var/len(self.array_y))
        
        if self.alternative == "neq":
            
            p_val = 2*(1-stats.norm.cdf(abs(t)))
            return p_val
        
        elif (self.alternative == ">"):
            
            p_val = 1-stats.norm.cdf(t)
            
            return p_val
        
        else:
            
            p_val = stats.norm.cdf(t)
            
            return p_val
        
    
    def f_test(self):
        
        samp1_var = np.var(self.array_x)*len(self.array_x)/(len(self.array_x)-1)
        samp2_var = np.var(self.array_y)*len(self.array_y)/(len(self.array_y)-1)
        
        f = samp1_var/samp2_var
        
        reject = f > stats.f.ppf(1-self.alpha/2,len(self.array_x)-1,len(self.array_y)-1) or f <stats.f.ppf(self.alpha/2,len(self.array_x)-1,len(self.array_y)-1)
        
        return reject
    
    
```

```python
np1 = np.random.normal(loc=10,scale=2,size=50).tolist()
np2 = np.random.normal(loc=7,scale=5,size=20).tolist()

c = Stats_Test(array_x=np1,array_y=np2,val=1,alternative=">",alpha=0.05)
c.test()
```
